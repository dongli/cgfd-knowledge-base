```python
# This example is taken from "Finite Difference Methods for Ordinary and Partial Differential Equations" by Randall J. LeVeque

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f1(u, t):
	return -np.sin(t)

def f2(u, t, c):
	return c * (u - np.cos(t)) - np.sin(t)

def euler_forward(f, u0, dt, tn):
	u = [u0]
	t = 0
	while t <= tn:
		u.append(u[-1] + dt * f(u[-1], t))
		t = t + dt
	return np.array(u)

u0 = 1
ut = np.cos(np.linspace(0, 2, u.shape[0]))

u1 = euler_forward(f1, u0, dt=1e-3, tn=2)

plt.subplot(321)
plt.plot(u1)
plt.gca().set_xlabel('Time')
plt.gca().set_ylabel('Solution')
plt.subplot(322)
plt.plot(u1 - ut)
plt.gca().set_xlabel('Time')
plt.gca().set_ylabel('Error')

u2 = euler_forward(lambda u, t: f2(u, t, c=-10), u0, dt=1e-3, tn=2)

plt.subplot(323)
plt.plot(u2)
plt.gca().set_xlabel('Time')
plt.gca().set_ylabel('Solution')
plt.subplot(324)
plt.plot(u2 - ut)
plt.gca().set_xlabel('Time')
plt.gca().set_ylabel('Error')

u3 = euler_forward(lambda u, t: f2(u, t, c=-2100), u0, dt=1e-3, tn=2)

plt.subplot(325)
plt.plot(u3)
plt.gca().set_xlabel('Time')
plt.gca().set_ylabel('Solution')
plt.subplot(326)
plt.plot(u3 - ut)
plt.gca().set_xlabel('Time')
plt.gca().set_ylabel('Error')

plt.gcf().set_size_inches(10, 10)
plt.subplots_adjust(wspace=0.6, hspace=0.4)
```
